* Na	 Presencia de Sodio en el Aceite
* Visco: Viscosidad del aceite
* Fecha de Análisis: Indica cuando fué analizada la muestra por el laboratorio

```python
values_engine.head()
```

**Clasificacion**

```python
#Imports
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV, train_test_split

#from ml.visualization import plot_confusion_matrix, plot_learning_curve

np.random.seed(42)
```

```python
y = values_engine['Resultado']
X = values_engine.drop('Resultado', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

**SGDClassifier**

```python
#from pactools.grid_search import GridSearchCVProgressBar
from ml.visualization import plot_confusion_matrix, plot_learning_curve


plt.figure(figsize=(10, 10), dpi= 80, facecolor='w', edgecolor='k')

for idx, loss in enumerate(('hinge', 'log', 'perceptron', 'modified_huber', 'squared_hinge'), start=1):
    exploring_params = {
        'eta0': [0.1, 0.01],  # Tasa de entrenamiento
        'alpha': [0.01, 0.001, 0.0001],  # Tasa de regularización
        'max_iter': [500, 1000, 2000],
        'learning_rate': ['optimal', 'invscaling', 'constant', 'adaptive'],
        'n_iter_no_change': [5, 7]
    }
    m = SGDClassifier(loss=loss, class_weight=None, verbose=0, n_jobs=-1, random_state=42, 
                     penalty='l2',
                     tol= 0.001,
                     epsilon= 0.1,
                     power_t=0.5,
                     validation_fraction=0.1,
                     fit_intercept=True,
                     average=False,
                     warm_start=False,
                     shuffle=True,
                     early_stopping=False
                     ) 
    model = GridSearchCV(m, exploring_params, cv=5, scoring='accuracy', verbose=0)
    model.fit(X_train, y_train)

    results = pd.DataFrame(columns=('clf', 'best_acc'))
    best_model = model.best_estimator_
    results = results.append({'clf': best_model, 'best_acc':  model.best_score_}, ignore_index=True)
    print('Best SGDClassifier accuracy: ', model.best_score_, 'for loss', loss)
    print('The best classifier so far is: ')
    print(results.loc[results['best_acc'].idxmax()]['clf'])

    print("Reporte de clasificación para el mejor clasificador (sobre conjunto de evaluación): ", end="\n\n")
    y_true, y_pred = y_test, model.predict(X_test)
    print(classification_report(y_true, y_pred), end="\n\n")

    print("================================================", end="\n\n")

    plt.subplot(3, 2, idx)
    plot_confusion_matrix(confusion_matrix(y_true, y_pred),
                          classes=['1', '2', '3'], title="Matriz de confusión para %s" % loss)
```

```python
 plot_confusion_matrix(confusion_matrix(y_true, y_pred),
                          classes=['1', '2', '3'], title="Matriz de confusión para %s" % 'squared_hinge')
```

**Decision Tree Classifier**

```python
#from pactools.grid_search import GridSearchCVProgressBar
from sklearn.tree import DecisionTreeClassifier

plt.figure(figsize=(14, 4), dpi= 80, facecolor='w', edgecolor='k')


exploring_params = {'criterion': ('entropy', 'gini'),
                    'max_depth': (14, 28, 50),
                    'max_leaf_nodes': (52, 92, 132),
                    'max_features': (10, 11, 12, 13, 14),
                    'min_samples_split': (20, 30, 50)

}
dt = DecisionTreeClassifier(splitter='best', 
                           min_samples_leaf=1, 
                           min_weight_fraction_leaf=0.0, 
                           random_state=42, 
                           min_impurity_decrease=0.0, 
                           min_impurity_split=None, 
                           class_weight=None, 
                           presort=False)
model_dt = GridSearchCV(dt, exploring_params, cv=5, scoring='accuracy', verbose=0)
model_dt.fit(X_train, y_train)

results = pd.DataFrame(columns=('clf', 'best_acc'))
best_model = model_dt.best_estimator_
results = results.append({'clf': best_model, 'best_acc':  model_dt.best_score_}, ignore_index=True)
print('Best Decision Tree accuracy: ', model_dt.best_score_)
print('The best classifier so far is: ')
print(results.loc[results['best_acc'].idxmax()]['clf'])

print("Reporte de clasificación para el mejor clasificador (sobre conjunto de evaluación): ", end="\n\n")
y_true, y_pred = y_test, model_dt.predict(X_test)
print(classification_report(y_true, y_pred), end="\n\n")

print("================================================", end="\n\n")

plt.subplot(1, 1, 1)
plot_confusion_matrix(confusion_matrix(y_true, y_pred),
                      classes=['1', '2', '3'], title="Matriz de confusión")
```

Decision Tree Classifier parece dar mejores resultados entre estos dos clasificadores
